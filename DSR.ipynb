```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('test_data.csv',encoding='iso-8859-1')
```

```python
df = df.set_index('Treshold for LR (default=0.5)')
fig=plt.figure()
ax=fig.add_axes([0,0,2.2,1.8])
ax.set_xlabel('treshold')
ax.set_ylabel('%')
ax.set_title('Accuracy, Precision, Recall')
ax.set_xticks([0,0.2,0.4,0.6,0.8,1])
ax.set_xticklabels([0,0.2,0.4,0.6,0.8,1])
df['Accuracy [%]'].plot()
df['Class precision 1 [%]'].plot()
df['Class recall 1 [%]'].plot()
ax.legend(loc='upper right', bbox_to_anchor=(1, 0.5))
```
